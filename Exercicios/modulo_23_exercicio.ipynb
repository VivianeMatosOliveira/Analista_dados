### **1.2. Variáveis numéricas** 

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
data[['comprimento_bico_mm','profundidade_bico_mm','comprimento_nadadeira_mm','massa_corporal_g']].describe().T
```

```
# resposta da questão 1.2
```

```
media = data['comprimento_bico_mm'].mean()
print(media)

desvio_padrao = data['comprimento_bico_mm'].std()
print(desvio_padrao)
```

```
data['comprimento_bico_mm_std'] = data['comprimento_bico_mm'].apply(lambda comprimento: (comprimento - media) / desvio_padrao)
```

```
media = data['comprimento_bico_mm_std'].mean()
print(media)

desvio_padrao = data['comprimento_bico_mm_std'].std()
print(desvio_padrao)
```

```
media = data['profundidade_bico_mm'].mean()
print(media)

desvio_padrao = data['profundidade_bico_mm'].std()
print(desvio_padrao)
```

```
data['profundidade_bico_mm_std'] = data['profundidade_bico_mm'].apply(lambda profundidade: (profundidade - media) / desvio_padrao)
```

```
media = data['comprimento_nadadeira_mm'].mean()
print(media)

desvio_padrao = data['comprimento_nadadeira_mm'].std()
print(desvio_padrao)
```

```
data['comprimento_nadadeira_mm_std'] = data['comprimento_nadadeira_mm'].apply(lambda comprimento: (comprimento - media) / desvio_padrao)
```

```
media = data['massa_corporal_g'].mean()
print(media)

desvio_padrao = data['massa_corporal_g'].std()
print(desvio_padrao)
```

```
data['massa_corporal_g_std'] = data['massa_corporal_g'].apply(lambda massa: (massa - media) / desvio_padrao)
```

### **1.3. Variáveis categóricas** 

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.3
```

```
data['sex_m_nom'] = data['sexo'].apply(lambda sexo: 1 if sexo == 'Male' else 0)
data['sex_f_nom'] = data['sexo'].apply(lambda sexo: 1 if sexo == 'Female' else 0)
```

```
ilha_mapper = {
   'Torgersen':1, 
   'Biscoe':2, 
   'Dream': 3,
}
```

```
data['ilha_encoded_ord'] = data['ilha'].apply(lambda level: ilha_mapper[level])
```
